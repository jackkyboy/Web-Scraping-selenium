```python
import nest_asyncio
import asyncio
from requests_html import AsyncHTMLSession

# Apply the nest_asyncio
nest_asyncio.apply()

async def fetch_data():
    # สร้างเซสชันใหม่
    session = AsyncHTMLSession()
    
    # ดึงข้อมูลจาก URL
    response = await session.get('https://www.tqm.co.th/car-insurance/all')
    
    # เรนเดอร์หน้าเว็บเพื่อให้ JavaScript ทำงาน
    await response.html.arender()
    
    # ตัวอย่าง: ดึงข้อมูล title ของหน้าเว็บ
    title = response.html.find('title', first=True).text
    print('Page Title:', title)
    
    # ตัวอย่าง: ดึงชื่อแผนประกันทั้งหมด
    # ปรับแต่ง selector ให้ตรงกับ HTML โครงสร้างของเว็บเพจที่คุณต้องการดึงข้อมูล
    insurance_plans = response.html.find('.plan-name')  # ปรับ .plan-name ให้ตรงกับ class ที่ต้องการ
    
    for plan in insurance_plans:
        print('Insurance Plan:', plan.text)

# เรียกใช้งานฟังก์ชัน fetch_data ใน event loop
asyncio.run(fetch_data())
```
